```python
import numpy as np
import matplotlib.pyplot as plt
import re

plt.style.use('ggplot')

sol_file = "izlaz-dec2"
obj_file = "izlaz-obj2"

with open(obj_file) as f:
    lines = f.readlines()
    x = [float(line[1:-2].split(",")[0]) for line in lines]
    y = [float(line[1:-2].split(",")[1]) for line in lines]

fig = plt.figure(figsize=(11, 7)) 
ax1 = fig.add_subplot()
plt.ylim(0, 10)
plt.xlim(0.1, 1)
ax1.set_xlabel('$f_1$', fontsize=18)
ax1.set_ylabel('$f_2$', fontsize=18)
plt.scatter(x, y)

plt.show()
```
